```python
# Define empty inventory list
inventory = []
# Define empty cart list
cart = []

# Function to upload inventory
def upload_inventory():
    item_name = input("Enter item name: ")
    item_code = input("Enter item code: ")
    price = float(input("Enter price: "))
    quantity = int(input("Enter quantity: "))
    inventory.append({'name': item_name, 'code': item_code, 'price': price, 'quantity': quantity})
    print("Inventory updated successfully.")

# Function to delete item from inventory
def delete_item():
    item_code = input("Enter item code to delete: ")
    for item in inventory:
        if item['code'] == item_code:
            inventory.remove(item)
            print(f"Item '{item['name']}' deleted successfully.")
            break
    else:
        print("Item not found in inventory.")

# Function to view inventory
def view_inventory():
    if inventory:
        print("Inventory:")
        print("Item Name\tItem Code\tPrice\tQuantity")
        for item in inventory:
            print(f"{item['name']}\t\t{item['code']}\t\t${item['price']}\t\t{item['quantity']}")
    else:
        print("Inventory is empty.")

# Function to generate bill
def generate_bill():
    if inventory:
        total_price = sum(item['price'] * item['quantity'] for item in inventory)
        print("Bill:")
        print("Item Name\t\tItem Code\t\tPrice\t\tQuantity\t\tTotal Price")
        for item in inventory:
            print(f"{item['name']}\t\t{item['code']}\t\t${item['price']}\t\t{item['quantity']}\t\t${item['price'] * item['quantity']}")
        print("Total Bill Amount:", total_price)
    else:
        print("Inventory is empty. Cannot generate bill.")

# Function for consumer to add items to cart
def add_to_cart():
    item_name = input("Enter item name to add to cart: ")
    quantity = int(input("Enter quantity: "))
    for item in inventory:
        if item['name'] == item_name:
            if item['quantity'] >= quantity:
                cart.append({'name': item['name'], 'code': item['code'], 'price': item['price'], 'quantity': quantity})
                print("Item added to cart successfully.")
                return True  # Return True if item added to cart successfully
            else:
                print("Not enough quantity available in the inventory.")
                return False  # Return False if item not added to cart due to insufficient quantity
    else:
        print("Item not found in inventory.")
        return False  # Return False if item not found in inventory

# Function to generate total bill for purchased items
def generate_total_bill():
    if cart:
        total_price = sum(item['price'] * item['quantity'] for item in cart)
        print("Total Bill Amount:", total_price)
    else:
        print("Your cart is empty.")

# Function to generate bill for items in the cart
def generate_cart_bill():
    if cart:
        print("Cart Bill:")
        print("Item Name\tItem Code\tPrice\tQuantity\tTotal Price")
        for item in cart:
            print(f"{item['name']}\t\t{item['code']}\t\t${item['price']}\t\t{item['quantity']}\t\t${item['price'] * item['quantity']}")
        total_price = sum(item['price'] * item['quantity'] for item in cart)
        print("Total Cart Bill Amount:", total_price)
    else:
        print("Your cart is empty.")

# Main program
while True:
    print("\nWelcome to the Supermarket Management System")
    print("1. Producer")
    print("2. Consumer")
    print("3. Exit")
    choice = input("Enter your choice: ")

    if choice == '1':
        print("\nProducer Menu:")
        print("1. Upload Inventory")
        print("2. Delete Item")
        print("3. View Inventory")
        print("4. Generate Bill")
        producer_choice = input("Enter your choice: ")

        if producer_choice == '1':
            upload_inventory()

        elif producer_choice == '2':
            delete_item()

        elif producer_choice == '3':
            view_inventory()

        elif producer_choice == '4':
            generate_bill()

        else:
            print("Invalid choice. Please try again.")

    elif choice == '2':
        print("\nConsumer Menu:")
        print("1. View Inventory")
        print("2. Add Item to Cart")
        print("3. Generate Cart Bill")
        print("4. Purchase")
        consumer_choice = input("Enter your choice: ")

        if consumer_choice == '1':
            view_inventory()

        elif consumer_choice == '2':
            add_to_cart()

        elif consumer_choice == '3':
            generate_cart_bill()

        elif consumer_choice == '4':
            generate_total_bill()
            inventory.clear()  # Clear inventory after purchase
            cart.clear()  # Clear cart after purchase
            print("Thank you for your purchase!")

        else:
            print("Invalid choice. Please try again.")

    elif choice == '3':
        print("Thank you for using the Supermarket Management System.")
        break

    else:
        print("Invalid choice. Please try again.")
```
